```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Create an ImageDataGenerator instance with desired augmentation parameters
datagen = ImageDataGenerator(
    # Add your desired augmentation parameters here (e.g., rotation, shear, zoom, etc.)
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)

# Load the original images from the directory
img_generator = datagen.flow_from_directory(
    directory='frs_data_crop',
    batch_size=32,
    class_mode='categorical',
    save_to_dir='frs_data_aug',
    save_format='jpeg'
)

# Set the total number of augmented images you want
total_augmented_images_needed = 1000
output_dir = "/home/janardhan/Documents/machine_learning/ML projects/FRS/model/frs_data_aug"
# Calculate the number of batches
num_batches = (total_augmented_images_needed // batch_size) + (1 if total_augmented_images_needed % batch_size != 0 else 0)

# Create a directory for each class to store augmented images
class_names = img_generator.class_indices.keys()
for class_name in class_names:
    os.makedirs(os.path.join(output_dir, class_name), exist_ok=True)

# Iterate through batches and save augmented images to the appropriate class folders
for i in range(num_batches):
    batch_images, batch_labels = next(img_generator)
    for j in range(len(batch_images)):
        image_path = img_generator.filenames[i * batch_size + j]
        class_label = image_path.split('/')[-2]  # Extract class name from image path
        augmented_image_path = os.path.join('frs_data_aug', class_label, image_path.split('/')[-1])
        img_generator.save_to_dir(augmented_image_path)
```
